## Bar plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,3,4,2]
c=['g','b','r','y']
plt.bar(x,y,color=c,width=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['English','Hindi',"python",'c','CA','DS']
y=[15,19,10,18,20,20]
plt.xlabel('subjects',fontsize=15)            # xlabel() a functions to set a label for the x-axis.
plt.ylabel('Marks',fontsize=15)               # ylabel() a functions to set a label for the y-axis.
plt.title('result',loc='left',fontsize=15)    # we can use the title() function to set a title for the plot.
                                              # loc parameter in title() to position the title.
c=['g','r','y','b','y','r']
plt.bar(x,y,width=0.4,color=c,edgecolor='g',linewidth=2,linestyle='-',alpha=0.7,label='Label',)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['English','Hindi',"python",'c','CA','DS']
y=[15,19,10,18,20,20]
z=[12,14,8,15,14,16]

width=0.3
p=np.arange(len(x))
p1=[j+ width for j  in p]

plt.xlabel('subjects',fontsize=15)
plt.ylabel('Marks',fontsize=15)
plt.title('result',fontsize=15)
c=['g','r','y','b','y','r']

plt.bar(p,y,width,color='r',label='Label')
plt.bar(p1,z,width,color='y',label='Label')

plt.xticks(p+width/2,x,rotation=15)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['English','Hindi',"python",'c','CA','DS']
y=[15,19,10,18,20,20]
z=[12,14,8,15,14,16]

width=0.3
p=np.arange(len(x))
p1=[j+ width for j  in p]

plt.xlabel('subjects',fontsize=15)
plt.ylabel('Marks',fontsize=15)
plt.title('result',fontsize=15)


plt.barh(p,y,width,color='r',label='Label',edgecolor='b')
plt.barh(p1,z,width,color='y',label='Label',edgecolor='b')

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

## Scatter plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[4,3,9,5,5,7,8]
c=['g','r','b','y','b','r','y']
sizes=[100,200,400,500,200,600,400]
plt.scatter(x,y,color=c,s=sizes,marker='*',edgecolor='g',linewidth=2)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=20)
plt.ylabel("Number",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[4,3,9,5,5,7,8]
y1=[3,6,7,4,5,6,7]
#c=['g','r','b','y','b','r','y']
sizes=[100,200,400,500,200,600,400]
plt.scatter(x,y,s=sizes,cmap='Dark2')

plt.scatter(x,y1,s=sizes,cmap='viridis')
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.show()
```

## subplot --> subplots() function in Python simplifies the creation of multiple subplots Matplotlib within a single figure, allowing for organized and simultaneous visualization of various datasets or plots.<br>

***The subplot() function takes three arguments that describes the layout of the figure.<br>

The layout is organized in rows and columns, which are represented by the first and second argument.<br>

The third argument represents the index of the current plot.***

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,3,4,2]
c=['g','b','r','y']
plt.subplot(2,2,1)
plt.title("Bar graph",fontsize=8)
plt.bar(x,y,color=c,width=0.2)

x1=[1,2,5,6]
y1=[4,6,7,9]
plt.subplot(2,2,2)
plt.title("scatter graph",fontsize=8)
plt.scatter(x1,y1)

x2=[1,2,5,6]
y2=[4,6,7,9]
plt.subplot(2,2,3)
plt.title("line graph",fontsize=8)
plt.plot(x2,y2)

x3=[1,2,5,6]
y3=[4,6,7,9]
plt.subplot(2,2,4)
plt.title("Horizontal Bar graph",fontsize=8)
plt.barh(x3,y3)

plt.show()

```

## Markers
***You can use the keyword argument marker to emphasize each point with a specified marker.***

## Marker Size
***You can use the keyword argument markersize or the shorter version, ms to set the size of the markers.***

## Marker Color
***You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers.***



```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'H', ms = 20, mec = 'r')
plt.show()
```

## Linestyle
***You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line.***<br>

### Line Width
***You can use the keyword argument linewidth or the shorter lw to change the width of the line.<br>

The value is a floating number, in points.***

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '7.5',linestyle='--')
plt.show()
```

## legends
**A legend is an area describing the elements of the graph. In the Matplotlib library, thereâ€™s a function called legend() which is used to place a legend on the axes.**

```python
import numpy as np 
import matplotlib.pyplot as plt 

# X-axis values 
x = [1, 2, 3, 4, 5] 

# Y-axis values 
y = [1, 4, 9, 16, 25] 

# Function to plot 
plt.plot(x, y) 

# Function add a legend 
plt.legend(['single element']) 

# function to show the plot 
plt.show() 
```

## Ticks
**To add ticks to a plot, you can use the xticks() and yticks() functions. These functions allow you to set the tick locations and labels for the x and y axes.**

```python
import matplotlib.pyplot as plt
import numpy as np
x=[7,5,9,8]
y=[1,4,7,9]

# Plot the data
plt.plot(x, y)

# Set the tick locations and labels
plt.xticks(np.arange(0, 11, 2))
plt.yticks(np.arange(1,10, 2))

# Show the plot
plt.show()
```
