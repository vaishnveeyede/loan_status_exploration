## Seaborn is a Python library which is based on matplotlib and is used for data visualization. It provides a medium to present data in a statistical graph format as an informative and attractive medium to impart some information.

## Single Line Plot
**A single line plot presents data on x-y axis using a line joining datapoints. To obtain a graph Seaborn comes with an inbuilt function to draw a line plot called lineplot().** 


**Use case: Useful for showing the progression of data points and identifying trends or patterns.**

**Seaborn Function: sns.lineplot()**


Syntax: lineplot(x,y,data)<br>

where,<br>

xâ€“ data variable for x-axis<br>

y- data variable for y-axis<br>

data- data to be plotted<br>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

d1={"name":['aditi','vaishu','jay','satyam','aashu','umang','shivani','sheetal'],
    'sex':['F','F','M','M','M','F','F','F'],
    "age":[10,20,22,15,18,20,23,24],
    "marks":[90,80,76,65,54,40,68,80],
    "city":['lanji','goa','delhi','lanji','goa','delhi','aagra','lanji']
   }
df=pd.DataFrame(d1)
print(df)

sns.lineplot(x='name',y='marks',data=df)
plt.show()
```

## Bar plot 
**Definition: A bar plot represents categorical data with rectangular bars. The length of each bar corresponds to the value of the data.**

**Use case: Suitable for comparing the values of different categories, showing the distribution or composition of data.**

**Seaborn Function: sns.barplot()**


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

d1={"name":['aditi','vaishu','jay','satyam','aashu','umang','shivani','sheetal'],
    'sex':['F','F','M','M','M','F','F','F'],
    "age":[10,20,22,15,18,20,23,24],
    "marks":[90,80,76,65,54,40,68,80],
    "city":['lanji','goa','delhi','lanji','goa','delhi','aagra','lanji']
   }
df=pd.DataFrame(d1)
print(df)

sns.barplot(x='name',y='marks',data=df,width=0.4)
plt.show()


```

## scatter plot


**Definition: A scatter plot is a two-dimensional data visualization that uses individual data points to represent the values of two variables.**

**Use case: It is used to visually examine the relationship between two continuous variables, helping identify patterns or trends in the data.**

**Seaborn Function: sns.scatterplot()**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

d1={"name":['aditi','vaishu','jay','satyam','aashu','umang','shivani','sheetal'],
    'sex':['F','F','M','M','M','F','F','F'],
    "age":[10,20,22,15,18,20,23,24],
    "marks":[90,80,76,65,54,40,68,80],
    "city":['lanji','goa','delhi','lanji','goa','delhi','aagra','lanji']
   }
df=pd.DataFrame(d1)
print(df)

sns.scatterplot(x='name',y='marks',data=df,marker='.',edgecolor='r',color='g')
plt.show()


```

## Histogram plot

**Definition: A histogram is a graphical representation of the distribution of a continuous dataset. It is created by dividing the data into bins and counting the number of observations in each bin.**

**Use case: Useful for understanding the underlying frequency distribution of a continuous variable.**

**Seaborn Function: sns.histplot()**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

d1={"name":['aditi','vaishu','jay','satyam','aashu','umang','shivani','sheetal'],
    'sex':['F','F','M','M','M','F','F','F'],
    "age":[10,20,30,40,50,60,70,80],
    "marks":[90,80,76,65,54,40,68,50],
    "city":['lanji','goa','delhi','lanji','goa','delhi','aagra','lanji']
   }
df=pd.DataFrame(d1)
print(df)

sns.displot(df['age'],bins=[10,20,30,40,50,60,70,80],kde=True, color='g',rug=True)
plt.show()
```

## Density plot

**a density plot using the sns.kdeplot() function. A density plot visualizes the distribution of a continuous variable and is particularly useful for understanding the shape of the data distribution.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('iris')
sns.kdeplot(data=data, x='sepal_length', fill=True)
plt.title('Density Plot of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.show()
```

## BOX PLOT

### A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. 

```python
data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Value': [4.5, 5.0, 4.8, 6.1, 6.5, 6.3, 7.2, 7.5, 7.1]
}
df = pd.DataFrame(data)
sns.boxplot(data=df, x='Category', y='Value')
plt.title('Box Plot of Values by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```
